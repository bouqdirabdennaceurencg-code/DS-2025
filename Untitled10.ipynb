```python
# --- [1] Installer les dépendances ---
!pip install kagglehub[pandas-datasets] pandas matplotlib seaborn

# --- [2] Importer les librairies ---
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

# --- [3] Charger le dataset ---
# Exemple de fichier : "WorldCups.csv" ou "WorldCupMatches.csv"
file_path = "WorldCups.csv"   # tu peux changer selon le fichier voulu

df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "abecklas/fifa-world-cup",
    file_path
)

print("Aperçu des 5 premières lignes :")
print(df.head())

# --- [4] Générer un compte rendu Markdown ---
report_lines = []
report_lines.append("# Compte rendu FIFA World Cup Dataset")
report_lines.append(f"**Date de génération :** {datetime.datetime.utcnow().isoformat()} UTC")
report_lines.append(f"**Fichier analysé :** {file_path}")
report_lines.append("")

# Statistiques générales
report_lines.append("## Statistiques générales")
report_lines.append(f"- Nombre de lignes : {len(df)}")
report_lines.append(f"- Nombre de colonnes : {len(df.columns)}")
report_lines.append("")

# Colonnes disponibles
report_lines.append("## Colonnes disponibles")
for col in df.columns:
    report_lines.append(f"- {col}")
report_lines.append("")

# Exemple d'analyse : Moyenne des buts et des spectateurs si colonnes présentes
if "GoalsScored" in df.columns:
    report_lines.append("## Analyse des buts")
    report_lines.append(f"- Moyenne des buts par édition : {df['GoalsScored'].mean():.2f}")
    report_lines.append(f"- Maximum des buts : {df['GoalsScored'].max()}")
    report_lines.append("")

if "Attendance" in df.columns:
    report_lines.append("## Analyse des spectateurs")
    report_lines.append(f"- Moyenne des spectateurs : {df['Attendance'].mean():.0f}")
    report_lines.append(f"- Maximum des spectateurs : {df['Attendance'].max()}")
    report_lines.append("")

# Sauvegarder le rapport
REPORT_PATH = "/content/FIFA_World_Cup_Report.md"
with open(REPORT_PATH, "w", encoding="utf-8") as f:
    f.write("\n".join(report_lines))

print(f"✅ Rapport Markdown généré : {REPORT_PATH}")

# --- [5] Afficher le rapport directement dans Colab ---
from IPython.display import Markdown, display
display(Markdown(open(REPORT_PATH, "r", encoding="utf-8").read()))

# --- [6] Exemple de visualisation ---
if "Year" in df.columns and "GoalsScored" in df.columns:
    plt.figure(figsize=(10,5))
    sns.lineplot(data=df.sort_values("Year"), x="Year", y="GoalsScored")
    plt.title("Nombre de buts par édition de la Coupe du Monde")
    plt.xlabel("Année")
    plt.ylabel("Buts")
    plt.grid(True)
    plt.show()
```
