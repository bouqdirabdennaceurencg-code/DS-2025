```python
# ============================================================
# Notebook complet : Implémentation Technique (Python)
# Dataset : Estimation of Obesity Levels (UCI id=544)
# ============================================================

# ------------------------------------------------------------
# 0. Install + Imports
# ------------------------------------------------------------
!pip install -q ucimlrepo

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from ucimlrepo import fetch_ucirepo

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, RandomizedSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_auc_score
)

from IPython.display import display

%matplotlib inline
sns.set(style="whitegrid")

# ------------------------------------------------------------
# 1. Chargement du dataset
# ------------------------------------------------------------
obesity = fetch_ucirepo(id=544)

X = obesity.data.features   # DataFrame des features
y = obesity.data.targets    # DataFrame/Series des cibles

print("Metadata du dataset :")
print(obesity.metadata)
print("\nVariables :")
print(obesity.variables)

# Fusion X + y pour l'EDA
if isinstance(y, pd.DataFrame):
    target_col = y.columns[0]
    df = X.copy()
    df[target_col] = y.iloc[:, 0]
else:
    target_col = "target"
    df = X.copy()
    df[target_col] = y

print("\nShape du dataset complet :", df.shape)
display(df.head())
display(df.info())

# ------------------------------------------------------------
# 2. Pré-traitement (Nettoyage, valeurs manquantes, doublons)
# ------------------------------------------------------------
data = df.copy()

# 2.1 Doublons
n_duplicates = data.duplicated().sum()
print(f"\nNombre de lignes dupliquées : {n_duplicates}")
if n_duplicates > 0:
    data = data.drop_duplicates()
    print("Doublons supprimés. Nouvelle shape :", data.shape)

# 2.2 Valeurs manquantes
print("\nValeurs manquantes par variable :")
missing = data.isna().sum().sort_values(ascending=False)
print(missing)

# 2.3 Séparation des types
numeric_cols = data.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = data.select_dtypes(include=["object", "category", "bool"]).columns.tolist()

print("\nVariables numériques :", numeric_cols)
print("Variables catégorielles :", categorical_cols)

# ------------------------------------------------------------
# 3. Feature Engineering (création de nouvelles variables)
# ------------------------------------------------------------
data_fe = data.copy()

# Exemple générique : score de "niveau de risque" basé sur les variables numériques
# -> nombre de variables numériques au-dessus de leur médiane (par individu)
if len(numeric_cols) > 0:
    medians = data_fe[numeric_cols].median()
    risk_flags = (data_fe[numeric_cols] > medians).astype(int)
    data_fe["numeric_risk_score"] = risk_flags.sum(axis=1)
    print("\nVariable de feature engineering ajoutée : 'numeric_risk_score'")

# ------------------------------------------------------------
# 4. Analyse Exploratoire des Données (EDA)
# ------------------------------------------------------------

eda = data_fe.copy()

# 4.1 Distribution des variables numériques
if len(numeric_cols) > 0:
    plt.figure(figsize=(16, 10))
    for i, col in enumerate(numeric_cols[:6], 1):  # max 6 graphes pour éviter surcharge
        plt.subplot(2, 3, i)
        sns.histplot(eda[col], bins=30, kde=True)
        plt.title(f"Distribution de {col}")
    plt.tight_layout()
    plt.show()

# 4.2 Distribution de la variable cible
plt.figure(figsize=(6, 4))
sns.countplot(x=eda[target_col])
plt.title("Répartition de la variable cible")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\nEffectifs par classe cible :")
print(eda[target_col].value_counts())

# 4.3 Boxplots pour quelques variables numériques vs cible
if len(numeric_cols) > 0:
    plt.figure(figsize=(16, 8))
    for i, col in enumerate(numeric_cols[:3], 1):
        plt.subplot(1, 3, i)
        sns.boxplot(x=eda[target_col], y=eda[col])
        plt.title(f"{col} selon la cible")
        plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# 4.4 Matrice de corrélation pour les variables numériques
if len(numeric_cols) > 1:
    corr_matrix = eda[numeric_cols].corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=False, cmap="coolwarm")
    plt.title("Matrice de corrélation (variables numériques)")
    plt.show()

# ------------------------------------------------------------
# 5. Préparation pour le Machine Learning
# ------------------------------------------------------------

ml_data = data_fe.copy()

# X / y
X_all = ml_data.drop(columns=[target_col])
y_all = ml_data[target_col].copy()

# Si la cible est catégorielle, on s'assure du type
y_all = y_all.astype("category")

# Redéfinir numeric / categorical sur X_all
numeric_features = X_all.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_features = X_all.select_dtypes(include=["object", "category", "bool"]).columns.tolist()

print("\nVariables numériques utilisées pour le ML :", numeric_features)
print("Variables catégorielles utilisées pour le ML :", categorical_features)

# Split train / test
X_train, X_test, y_train, y_test = train_test_split(
    X_all, y_all,
    test_size=0.2,
    random_state=42,
    stratify=y_all
)

print("\nTaille train :", X_train.shape)
print("Taille test  :", X_test.shape)

# ------------------------------------------------------------
# 6. Pipelines de pré-traitement (imputation, encodage, scaling)
# ------------------------------------------------------------

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

# ------------------------------------------------------------
# 7. Modélisation : test de 3 algorithmes + Cross-Validation
# ------------------------------------------------------------

models = {
    "LogisticRegression": LogisticRegression(
        max_iter=1000,
        multi_class="multinomial",
        n_jobs=-1
    ),
    "RandomForest": RandomForestClassifier(random_state=42),
    "GradientBoosting": GradientBoostingClassifier(random_state=42)
}

results = {}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("\n======= Comparaison des modèles (Accuracy CV) =======")
for name, clf in models.items():
    pipe = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("classifier", clf)
    ])

    cv_scores = cross_val_score(
        pipe,
        X_train, y_train,
        cv=cv,
        scoring="accuracy",
        n_jobs=-1
    )

    results[name] = cv_scores
    print(f"\nModèle : {name}")
    print("Scores Accuracy CV :", np.round(cv_scores, 4))
    print("Moyenne Accuracy   :", cv_scores.mean().round(4))
    print("Écart-type         :", cv_scores.std().round(4))

# ------------------------------------------------------------
# 8. Optimisation d’hyperparamètres (RandomForest + RandomizedSearchCV)
# ------------------------------------------------------------

rf_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(random_state=42))
])

param_distributions = {
    "classifier__n_estimators": [100, 200, 300],
    "classifier__max_depth": [None, 5, 10, 20],
    "classifier__min_samples_split": [2, 5, 10],
    "classifier__min_samples_leaf": [1, 2, 4],
    "classifier__max_features": ["sqrt", "log2"]
}

random_search = RandomizedSearchCV(
    estimator=rf_pipeline,
    param_distributions=param_distributions,
    n_iter=20,
    scoring="accuracy",
    cv=cv,
    verbose=2,
    n_jobs=-1,
    random_state=42
)

print("\n======= Lancement de la RandomizedSearchCV (RandomForest) =======")
random_search.fit(X_train, y_train)

print("\nMeilleurs hyperparamètres trouvés :")
print(random_search.best_params_)
print("Meilleure Accuracy moyenne en CV :", random_search.best_score_)

# ------------------------------------------------------------
# 9. Évaluation finale sur le test set
# ------------------------------------------------------------

best_model = random_search.best_estimator_

y_pred = best_model.predict(X_test)

print("\n======= Évaluation sur le set de test =======")
print("\nRapport de classification :")
print(classification_report(y_test, y_pred))

# Matrice de confusion
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(7, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_model.named_steps["classifier"].classes_)
disp.plot(cmap="Blues", xticks_rotation=45)
plt.title("Matrice de confusion - Meilleur modèle (RandomForest optimisé)")
plt.tight_layout()
plt.show()

# AUC multi-classes (si probabilités disponibles)
try:
    y_proba = best_model.predict_proba(X_test)
    # roc_auc_score en mode multi-classes (OVR)
    auc_macro = roc_auc_score(pd.get_dummies(y_test), y_proba, multi_class="ovr", average="macro")
    print(f"AUC-ROC macro (OVR) sur le test set : {auc_macro:.4f}")
except Exception as e:
    print("Impossible de calculer l'AUC multi-classes :", e)

# ------------------------------------------------------------
# Fin du notebook
# ------------------------------------------------------------
print("\nPipeline complet exécuté :")
print("- Pré-traitement (nettoyage, imputation, encodage, normalisation)")
print("- Analyse exploratoire (distributions, corrélations, boxplots)")
print("- Feature engineering (numeric_risk_score)")
print("- Modélisation avec 3 algorithmes et validation croisée")
print("- Optimisation d'hyperparamètres (RandomForest)")
print("- Évaluation finale sur le set de test")
```
