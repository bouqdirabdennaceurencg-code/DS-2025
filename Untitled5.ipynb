```python
# Installation des dépendances
!pip install -q kagglehub

# Import des bibliothèques
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd

# Chargement du dataset
print("Chargement du dataset en cours...")
try:
    df = kagglehub.load_dataset(
        KaggleDatasetAdapter.PANDAS,
        "emanfatima2025/college-placement-analysis"
    )
except:
    # Alternative si le premier chargement échoue
    path = kagglehub.dataset_download("emanfatima2025/college-placement-analysis")
    import os
    csv_file = [f for f in os.listdir(path) if f.endswith('.csv')][0]
    df = pd.read_csv(os.path.join(path, csv_file))

# Affichage des informations de base
print("\n" + "="*50)
print("INFORMATIONS SUR LE DATASET")
print("="*50)
print(f"\nNombre de lignes: {len(df)}")
print(f"Nombre de colonnes: {len(df.columns)}")

print("\n" + "="*50)
print("COLONNES DU DATASET")
print("="*50)
print(df.columns.tolist())

print("\n" + "="*50)
print("PREMIERS ENREGISTREMENTS")
print("="*50)
print(df.head())

print("\n" + "="*50)
print("INFORMATIONS SUR LES TYPES DE DONNÉES")
print("="*50)
print(df.info())

print("\n" + "="*50)
print("STATISTIQUES DESCRIPTIVES")
print("="*50)
print(df.describe())

print("\n" + "="*50)
print("VALEURS MANQUANTES")
print("="*50)
print(df.isnull().sum())

print("\n✓ Dataset chargé avec succès!")
```
