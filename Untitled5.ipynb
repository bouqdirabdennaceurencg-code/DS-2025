```python
# Installation des dÃ©pendances
!pip install -q kagglehub

# Import des bibliothÃ¨ques
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import display

# Configuration des graphiques
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

# Chargement du dataset
print("ðŸ“Š Chargement du dataset en cours...")
try:
    df = kagglehub.load_dataset(
        KaggleDatasetAdapter.PANDAS,
        "emanfatima2025/college-placement-analysis"
    )
except:
    path = kagglehub.dataset_download("emanfatima2025/college-placement-analysis")
    import os
    csv_file = [f for f in os.listdir(path) if f.endswith('.csv')][0]
    df = pd.read_csv(os.path.join(path, csv_file))

print(f"âœ“ Dataset chargÃ©: {len(df)} lignes, {len(df.columns)} colonnes\n")

# ==================== SECTION 1: VUE D'ENSEMBLE ====================
print("="*70)
print("ðŸ“‹ SECTION 1: VUE D'ENSEMBLE DU DATASET")
print("="*70)
print("\nðŸ”¹ AperÃ§u des premiÃ¨res lignes:")
display(df.head(10))

print("\nðŸ”¹ Informations sur les colonnes:")
display(pd.DataFrame({
    'Type': df.dtypes,
    'Non-Null': df.count(),
    'Null': df.isnull().sum(),
    '% Null': (df.isnull().sum() / len(df) * 100).round(2)
}))

print("\nðŸ”¹ Statistiques descriptives:")
display(df.describe().round(2))

# ==================== SECTION 2: ANALYSE DES PLACEMENTS ====================
print("\n" + "="*70)
print("ðŸŽ¯ SECTION 2: ANALYSE DES PLACEMENTS")
print("="*70)

# DÃ©tection automatique de la colonne de placement
placement_cols = [col for col in df.columns if 'place' in col.lower() or 'status' in col.lower()]
if placement_cols:
    placement_col = placement_cols[0]

    # Tableau de rÃ©partition des placements
    placement_counts = df[placement_col].value_counts()
    placement_pct = df[placement_col].value_counts(normalize=True) * 100
    placement_table = pd.DataFrame({
        'Nombre': placement_counts,
        'Pourcentage': placement_pct.round(2)
    })
    print(f"\nðŸ”¹ RÃ©partition des placements ({placement_col}):")
    display(placement_table)

    # Graphique: Distribution des placements
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Diagramme en barres
    placement_counts.plot(kind='bar', ax=axes[0], color='skyblue', edgecolor='black')
    axes[0].set_title('Distribution des Placements', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Statut de Placement')
    axes[0].set_ylabel('Nombre d\'Ã©tudiants')
    axes[0].tick_params(axis='x', rotation=45)

    # Diagramme circulaire
    colors = sns.color_palette('pastel')[0:len(placement_counts)]
    axes[1].pie(placement_counts, labels=placement_counts.index, autopct='%1.1f%%',
                colors=colors, startangle=90)
    axes[1].set_title('Proportion des Placements', fontsize=14, fontweight='bold')

    plt.tight_layout()
    plt.show()

# ==================== SECTION 3: ANALYSE PAR GENRE ====================
print("\n" + "="*70)
print("ðŸ‘¥ SECTION 3: ANALYSE PAR GENRE")
print("="*70)

gender_cols = [col for col in df.columns if 'gender' in col.lower() or 'sex' in col.lower()]
if gender_cols:
    gender_col = gender_cols[0]

    # Tableau de rÃ©partition par genre
    gender_table = pd.DataFrame({
        'Total': df[gender_col].value_counts(),
        'Pourcentage': (df[gender_col].value_counts(normalize=True) * 100).round(2)
    })
    print(f"\nðŸ”¹ RÃ©partition par genre ({gender_col}):")
    display(gender_table)

    # Si colonne de placement existe, analyse croisÃ©e
    if placement_cols:
        print(f"\nðŸ”¹ Placement par genre:")
        gender_placement = pd.crosstab(df[gender_col], df[placement_col],
                                       margins=True, margins_name='Total')
        display(gender_placement)

        # Graphique: Placement par genre
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))

        # Graphique groupÃ©
        pd.crosstab(df[gender_col], df[placement_col]).plot(kind='bar', ax=axes[0],
                                                             edgecolor='black')
        axes[0].set_title('Placements par Genre', fontsize=14, fontweight='bold')
        axes[0].set_xlabel('Genre')
        axes[0].set_ylabel('Nombre d\'Ã©tudiants')
        axes[0].legend(title='Statut')
        axes[0].tick_params(axis='x', rotation=0)

        # Graphique empilÃ© normalisÃ© (%)
        pd.crosstab(df[gender_col], df[placement_col], normalize='index').plot(
            kind='bar', stacked=True, ax=axes[1], edgecolor='black')
        axes[1].set_title('Taux de Placement par Genre (%)', fontsize=14, fontweight='bold')
        axes[1].set_xlabel('Genre')
        axes[1].set_ylabel('Proportion')
        axes[1].legend(title='Statut')
        axes[1].tick_params(axis='x', rotation=0)

        plt.tight_layout()
        plt.show()

# ==================== SECTION 4: ANALYSE DES NOTES/SCORES ====================
print("\n" + "="*70)
print("ðŸ“Š SECTION 4: ANALYSE DES PERFORMANCES ACADÃ‰MIQUES")
print("="*70)

# DÃ©tection des colonnes numÃ©riques (scores, notes, pourcentages)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
score_cols = [col for col in numeric_cols if any(x in col.lower() for x in
              ['score', 'percentage', 'cgpa', 'gpa', 'marks', 'grade'])]

if score_cols:
    print(f"\nðŸ”¹ Statistiques des scores:")
    stats_table = df[score_cols].describe().T
    stats_table['variance'] = df[score_cols].var()
    display(stats_table.round(2))

    # Distribution des scores
    n_scores = len(score_cols)
    if n_scores > 0:
        fig, axes = plt.subplots((n_scores + 1) // 2, 2, figsize=(14, 4 * ((n_scores + 1) // 2)))
        if n_scores == 1:
            axes = [axes]
        else:
            axes = axes.flatten()

        for idx, col in enumerate(score_cols[:len(axes)]):
            df[col].hist(bins=20, ax=axes[idx], color='lightgreen', edgecolor='black', alpha=0.7)
            axes[idx].set_title(f'Distribution: {col}', fontsize=12, fontweight='bold')
            axes[idx].set_xlabel('Score')
            axes[idx].set_ylabel('FrÃ©quence')
            axes[idx].axvline(df[col].mean(), color='red', linestyle='--',
                            label=f'Moyenne: {df[col].mean():.2f}')
            axes[idx].legend()

        # Cacher les axes vides
        for idx in range(len(score_cols), len(axes)):
            axes[idx].set_visible(False)

        plt.tight_layout()
        plt.show()

    # Boxplot des scores
    if len(score_cols) > 1:
        plt.figure(figsize=(12, 6))
        df[score_cols].boxplot()
        plt.title('Comparaison des Scores (Boxplot)', fontsize=14, fontweight='bold')
        plt.ylabel('Score')
        plt.xticks(rotation=45, ha='right')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

# ==================== SECTION 5: CORRÃ‰LATIONS ====================
print("\n" + "="*70)
print("ðŸ”— SECTION 5: ANALYSE DES CORRÃ‰LATIONS")
print("="*70)

if len(numeric_cols) > 1:
    # Matrice de corrÃ©lation
    corr_matrix = df[numeric_cols].corr()
    print("\nðŸ”¹ Matrice de corrÃ©lation:")
    display(corr_matrix.round(3))

    # Heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
                center=0, square=True, linewidths=1)
    plt.title('Heatmap des CorrÃ©lations', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

# ==================== SECTION 6: ANALYSE PAR SPÃ‰CIALISATION ====================
print("\n" + "="*70)
print("ðŸŽ“ SECTION 6: ANALYSE PAR SPÃ‰CIALISATION")
print("="*70)

spec_cols = [col for col in df.columns if any(x in col.lower() for x in
             ['special', 'stream', 'branch', 'degree', 'major', 'field'])]

if spec_cols:
    spec_col = spec_cols[0]

    # Tableau de rÃ©partition
    spec_counts = df[spec_col].value_counts()
    spec_table = pd.DataFrame({
        'Nombre': spec_counts,
        'Pourcentage': (spec_counts / len(df) * 100).round(2)
    })
    print(f"\nðŸ”¹ RÃ©partition par spÃ©cialisation ({spec_col}):")
    display(spec_table)

    # Graphique
    plt.figure(figsize=(12, 6))
    spec_counts.plot(kind='barh', color='coral', edgecolor='black')
    plt.title('Distribution des SpÃ©cialisations', fontsize=14, fontweight='bold')
    plt.xlabel('Nombre d\'Ã©tudiants')
    plt.ylabel('SpÃ©cialisation')
    plt.tight_layout()
    plt.show()

    # Analyse croisÃ©e avec placement si disponible
    if placement_cols:
        print(f"\nðŸ”¹ Placement par spÃ©cialisation:")
        spec_placement = pd.crosstab(df[spec_col], df[placement_col],
                                     margins=True, margins_name='Total')
        display(spec_placement)

        # Taux de placement par spÃ©cialisation
        placement_rate = pd.crosstab(df[spec_col], df[placement_col], normalize='index') * 100

        plt.figure(figsize=(12, 6))
        placement_rate.plot(kind='bar', stacked=True, edgecolor='black')
        plt.title('Taux de Placement par SpÃ©cialisation (%)', fontsize=14, fontweight='bold')
        plt.xlabel('SpÃ©cialisation')
        plt.ylabel('Pourcentage')
        plt.legend(title='Statut', bbox_to_anchor=(1.05, 1))
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

# ==================== SECTION 7: TOP INSIGHTS ====================
print("\n" + "="*70)
print("ðŸ’¡ SECTION 7: INSIGHTS CLÃ‰S")
print("="*70)

insights = []

# Insight 1: Taux de placement global
if placement_cols:
    placed = df[placement_col].value_counts()
    if len(placed) > 0:
        top_status = placed.index[0]
        top_pct = (placed.iloc[0] / len(df) * 100)
        insights.append(f"ðŸ“Œ {top_pct:.1f}% des Ã©tudiants ont le statut '{top_status}'")

# Insight 2: Genre dominant
if gender_cols:
    top_gender = df[gender_col].value_counts().index[0]
    top_gender_pct = (df[gender_col].value_counts().iloc[0] / len(df) * 100)
    insights.append(f"ðŸ“Œ {top_gender_pct:.1f}% des Ã©tudiants sont de genre '{top_gender}'")

# Insight 3: Meilleure performance
if score_cols:
    best_score_col = df[score_cols].mean().idxmax()
    best_avg = df[best_score_col].mean()
    insights.append(f"ðŸ“Œ Meilleure moyenne acadÃ©mique: {best_score_col} ({best_avg:.2f})")

# Insight 4: SpÃ©cialisation populaire
if spec_cols:
    top_spec = df[spec_col].value_counts().index[0]
    top_spec_count = df[spec_col].value_counts().iloc[0]
    insights.append(f"ðŸ“Œ SpÃ©cialisation la plus populaire: {top_spec} ({top_spec_count} Ã©tudiants)")

print("\n")
for insight in insights:
    print(insight)

print("\n" + "="*70)
print("âœ… ANALYSE TERMINÃ‰E")
print("="*70)
```
